```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
def show(image, cmap='gray'):
    plt.imshow(image, cmap=cmap)
    plt.show()
    
def export(imagename):
    cv2.imwrite(imagename+".jpg", eval(imagename))
    
temp_low  = 19.0
temp_high = 34.4
```

```python
img00 = cv2.imread("temp_scale.jpg")
img01 = cv2.cvtColor(img00, cv2.COLOR_BGR2RGB)
show(img01)
```

```python
img02 = np.mean(img01, axis=0).T
```

```python
temp = np.linspace(temp_low, temp_high, img02.shape[1])
plt.plot(temp, img02[0], c='r')
plt.plot(temp, img02[1], c='g')
plt.plot(temp, img02[2], c='b')
```

```python
red_c   = np.polyfit(temp, img02[0], 5)
green_c = np.polyfit(temp, img02[1], 5)
blue_c  = np.polyfit(temp, img02[2], 5)
red_p   = np.poly1d(red_c)
green_p = np.poly1d(green_c)
blue_p  = np.poly1d(blue_c)
```

```python
plt.plot(temp, red_p(temp),   c='r')
plt.plot(temp, green_p(temp), c='g')
plt.plot(temp, blue_p(temp),  c='b')
```
